### Handle Outliers

```python
def outlier_capping(data, feature):
    q1 = data[feature].quantile(0.25)
    q3 = data[feature].quantile(0.75)
    IQR = q3 - q1
    lower = q1 -1.5 * IQR
    upper = q3 + 1.5 * IQR

    # Count outliers before capping
    outliers = data[(data[feature] < lower) | (data[feature] > upper)].sum()
 
    if outliers > 0:
        # Cap outliers
        data[feature] = np.where(data[feature] < lower, lower, np.where(data[feature] > upper, upper))
    
    print(f'Number of outliers in {feature} before capping: {outliers}.shape')
    print(f'Number of outliers in {feature} after capping: {outliers}.shape')

    return data
```
